```python
for i in range(48):
    xlrd_book = xlrd.open_workbook("../"+carpeta[1]+"/"+sismo[i+1]+"."+extension[1], on_demand=True)
    with pd.ExcelFile(xlrd_book) as xlsx:
        # GRADO 3.6
        df1 = pd.read_excel(xlsx, grados[1])
        df1 = df1.drop(df1.columns[5],axis='columns')   #   Eliminar la columna 5
        my_df = pd.DataFrame(df1)                 #   Guardarlo en un csv en la carpeta data csv
        my_df.to_csv(sismo[i+1]+"  "+grados[1]+"."+extension[2], index=False, header=df1.columns,sep=';',decimal='.')
        # GRADO 5.3
        df2 = pd.read_excel(xlsx, grados[2])
        df2 = df2.drop(df2.columns[5],axis='columns')
        my_df = pd.DataFrame(df2) 
        my_df.to_csv(sismo[i+1]+"  "+grados[2]+"."+extension[2], index=False, header=df2.columns,sep=';',decimal='.')
        # GRADO 7.2
        df3 = pd.read_excel(xlsx, grados[3])
        df3 = df3.drop(df3.columns[5],axis='columns')
        my_df = pd.DataFrame(df3) 
        my_df.to_csv(sismo[i+1]+"  "+grados[3]+"."+extension[2], index=False, header=df3.columns,sep=';',decimal='.')
    
```
